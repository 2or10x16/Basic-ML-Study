### 차원 축소  

많은 feature로 구성된 다차원 데이터 세트의 차원을 축소해 새로운 차원의 데이터 세트 생성하는 것

#### 차원 축소를 하는 이유  

* 시각화  
* 노이즈 제거  
* 메모리 절약
* 퍼포먼스 향상

### 차원의 저주 


### 차원 축소 접근법

1) 투영  


학습 데이터셋은 고차원 공간에서 저차원 부분 공간에 수직으로 위치  
즉, 고차원의 데이터의 특성 중 일부 특성으로 데이터를 표현할 수 있다



2) 매니폴드 학습  



대부분의 차원 축소 알고리즘은 매니폴드를 모델링 학습하는 방식


차원 감소 시 학습 속도는 빨라지지만 항상 성능이 향상되는 것은 아님  
∴데이터셋이 어떤 모양을 하고 있냐에 따라 다름

### PCA  

여러 변수 간 상관관계를 이용해 이를 대표하는 주성분을 추출해 차원 축소

원본 데이터셋과 투영된 데이터셋 간의 평균제곱거리를 최소화 하는 축,즉 분산이 최대가 되는 축을 찾아서 이 축으로 차원 축소


#### 과정  
1)학습 데이터셋에서 분산이 최대인 축을 찾는다.   
2)이렇게 찾은 첫번째 축과 직교하면서 분산이 최대인 두 번째 축을 찾는다.  
3)첫 번째 축과 두 번째 축에 직교하고 분산을 최대한 보존하는 세 번째 축을 찾는다.  
4)1~3과 같은 방법으로 데이터셋의 차원(특성 수)만큼의 축을 찾는다.

### 군집  
 : 비슷한 샘플을 구별해 하나의 클러스터로 할당

* 군집의 목표 : 서로 유사한 데이터들을 같은 그룹으로 다르면 다른 그룹으로 분리

#### K-Means  


1)일단 K개의 임의의 중심점(centroid)을 배치하고  
2)각 데이터들을 가장 가까운 중심점으로 할당한다.  
3)군집으로 지정된 데이터들을 기반으로 해당 군집의 중심점을 업데이트한다.  
4)2번, 3번 단계를 그래서 수렴이 될 때까지, 즉 더이상 중심점이 업데이트 되지 않을 때까지 반복한다.

* 장점  

알고리즘이 쉽고 간결  

* 단점  

속성의 개수가 많으면 정확도가 떨어짐  

반복 횟수가 많으면 속도가 느려짐

#### DBSCAN  
:밀도 기반 군집화 알고리즘  

* 동일한 클래스에 속하는 데이터는 서로 근접하게 분포할 것이다"라는 가정을 기반으로 동작

* 장점  

기하학적으로 복잡한 데이터 세트도 군집화 가능   

이상치에 의해 클러스터 성능 감소하는 것 완화  

* 단점  

데이터 입력 순서, 거리 측정 방법에 따라 클러스터링 결과 변함
